```python
import pandas as pd
import numpy as np
import json
from datetime import datetime
```

```python
with open('weather_data_final.json', 'r') as json_file:
    data = json.load(json_file)
```

```python
df = pd.read_csv("data\\solar_panel_analysis_dataset_final.csv")
```

```python
df = df.drop(df.columns[:2], axis=1)
```

```python
ids_to_filter = data.keys()
```

```python
filtered_df = df[df['GEM phase ID'].isin(ids_to_filter)]
```

```python
id_list = filtered_df['GEM phase ID'].tolist()
```

```python


def create_dataframe_dict(id_list, data):
    progress = ""
    dataframe_dict = {}
    for phase_id in id_list:
        for freq in data[phase_id].keys():
            if freq != "full":
                for date in data[phase_id][freq].keys():
                    for param in data[phase_id][freq][date].keys():
                        for statistical,values in data[phase_id][freq][date][param].items():
                            key_name = "{}_{}_{}_{}".format(freq, param, statistical, date)
                            if  key_name not in dataframe_dict.keys():
                                dataframe_dict[key_name] = [values]
                            else:
                                dataframe_dict[key_name].append(values)
            else:
                for param in data[phase_id][freq].keys():
                    for statistical,values in data[phase_id][freq][param].items():
                        key_name = "{}_{}_{}".format(freq, param, statistical)
                        if  key_name not in dataframe_dict.keys():
                            dataframe_dict[key_name] = [values]
                        else:
                            dataframe_dict[key_name].append(values)
        current_progress = round((id_list.index(phase_id)/len(id_list)) * 100)
        if current_progress % 10 == 0 and current_progress != progress:
            print("{} - Progress: {}%".format(datetime.now(),current_progress))
            progress = current_progress
    return dataframe_dict


df_dictionary = create_dataframe_dict(id_list,data)
```

```python
for column in df_dictionary.keys():
    colum_series = pd.Series(df_dictionary[column])
    filtered_df[column] = colum_series
```
